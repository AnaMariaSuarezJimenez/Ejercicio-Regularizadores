Importamos las librerias necesarias:

```python
import random
import math
```

Definimos nuestra función objetivo:

```python
def f(x):
    return ((1 - (11*x - 7/2)**2) * math.cos(11*x - 7/2) + 1) + 2
```

Genera un individuo aleatorio en el dominio [0, 0.875] :

```python
def generar_individuo():
    return random.uniform(0, 0.875)
```

Selección de individuos:



```python
def seleccionar_padres(poblacion, valores_fitness, num_padres):
    padres = []
    for _ in range(num_padres):
        participantes = random.sample(list(range(len(poblacion))), 2)
        indice_ganador = participantes[0] if valores_fitness[participantes[0]] > valores_fitness[participantes[1]] else participantes[1]
        padres.append(poblacion[indice_ganador])
    return padres
```

Cruzamiento

```python
def cruzar(padre1, padre2):
    punto_corte = random.randint(1, len(padre1) - 1)
    hijo = padre1[:punto_corte] + padre2[punto_corte:]
    return hijo
```
